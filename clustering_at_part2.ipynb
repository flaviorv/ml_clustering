###Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_swiss_roll
from sklearn.cluster import DBSCAN
```

###DBSCAN with Make Swiss Roll dataset

```python
# Loading dataset Make Swiss Roll
x_msr, y_msr = make_swiss_roll(n_samples=1200, noise=0.5, random_state=42)

# Using DBSCAN and ploting scatter plots for many eps and min_samples values
eps = [0.5, 1, 2, 3]
min_samples = [2, 4, 6, 12]
fig, ax = plt.subplots(4,4, figsize=(18,15), subplot_kw={'projection':'3d'})
for i, e in enumerate(eps):
  for j, ms in enumerate(min_samples):
    dbscan = DBSCAN(eps=e, min_samples=ms)
    labels = dbscan.fit_predict(x_msr)
    ax[i, j].scatter(x_msr[:, 0], x_msr[:, 1], x_msr[:, 2], c=labels, cmap='Set1')
    clusters = np.unique(labels)
    has_noise = False
    if -1 in clusters:
      has_noise = True
    ax[i, j].set_title(f'eps: {e} min_samples: {ms}\nclusters: {clusters.shape[0]} has_noise: {has_noise}')
plt.suptitle('Make Swiss Roll Dataset with DBSCAN')
plt.tight_layout()
plt.show()
```
