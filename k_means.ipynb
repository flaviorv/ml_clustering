```python
import mglearn
mglearn.plots.plot_kmeans_algorithm()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#generating dataset
x, y = make_blobs(random_state=42)

#building cluster model
kmeans = KMeans(n_clusters=3)
#calculating centroids until they find the cluster center
kmeans.fit(x)

print('Labels assigned to data points:')
print(kmeans.labels_)

#ploting clusters vizualization with yours/theirs/other centroids
mglearn.discrete_scatter(x[:, 0], x[:, 1], kmeans.labels_, 'o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], [0,1,2], '^', c=['lightblue', 'gold', 'lightgreen'])
plt.legend(['Cluster 0','Cluster 1','Cluster 2', 'Cluster 0 center', 'Cluster 1 center', 'Cluster 2 center'])
plt.show()
```
